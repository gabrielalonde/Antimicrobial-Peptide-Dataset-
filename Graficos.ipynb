```python
import pandas as pd
import matplotlib.pyplot as plt
from Bio.SeqUtils.ProtParam import ProteinAnalysis

# Abre o arquivo e le a coluna
#grupo positivo#
positivo_df = pd.read_csv('grupo+.csv')
positivo_peptideos = positivo_df['sequencia'].tolist()

#grupo negativo#
negativo_df = pd.read_csv('grupo-.csv')
negativo_peptideos = negativo_df['sequencia'].tolist()

#Calcula a frequência de aminoácidos#
positivo_frequencias = [ProteinAnalysis(seq).get_amino_acids_percent() for seq in positivo_peptideos]
negativo_frequencias = [ProteinAnalysis(seq).get_amino_acids_percent() for seq in negativo_peptideos]

# Obter a lista de aminoácidos
aminoacidos = list(positivo_frequencias[0].keys())

# Calcular a média das frequências para cada aminoácido em cada grupo
frequencias_positivo = {aa: sum(freq[aa] for freq in positivo_frequencias) / len(positivo_frequencias) for aa in aminoacidos}
frequencias_negativo = {aa: sum(freq[aa] for freq in negativo_frequencias) / len(negativo_frequencias) for aa in aminoacidos}

# Cria um gráfico de barras comparativo
fig, ax = plt.subplots()
largura_barra = 0.35
indice = range(len(aminoacidos))

bar_positivo = ax.bar(indice, list(frequencias_positivo.values()), largura_barra, label='Positivo')
bar_negativo = ax.bar([i + largura_barra for i in indice], list(frequencias_negativo.values()), largura_barra, label='Negativo')

#eixo x
ax.set_xticks([i + largura_barra / 2 for i in indice])
ax.set_xticklabels(aminoacidos)
ax.set_xlabel('Aminoácidos')

#eixo y
ax.set_ylabel('Frequência')

ax.legend()

plt.show()
```

DISTRIBUIÇÃO DOS TAMANHOS DAS SEQUENCIAS DE PEPTPIDEOS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Carrega o arquivo CSV do grupo negativo
df = pd.read_csv('grupo-csv')
frequencia = df['tamanho'].value_counts()

sns.set(style="white") #estilo da gráfico

# Cria o gráfico
plt.figure(figsize=(10, 6))  # Tamanho da figura
ax = sns.barplot(x=frequencia.index, y=frequencia.values)
plt.xlabel('Tamanho da sequência do peptídeo') #rótulos
plt.ylabel('Frequência')
plt.title('Gráfico de Frequência')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
